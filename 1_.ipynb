# **Apple Data Visualisations**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# S closing price
plt.figure(figsize=(12, 5))
plt.plot(Apple_df["Date"], Apple_df["Close ($)"])
plt.title("Apple Closing Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

### Apple Daily Trading Volume Over Time

```python
plt.figure(figsize=(12, 5))
plt.plot(Apple_df["Date"], Apple_df["Volume"], color='orange')
plt.title("Apple Daily Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(True)
plt.show()
```

### Apple Daily Returns Distribution

```python
Apple_df['Daily Return'] = Apple_df['Close ($)'].pct_change() * 100

plt.figure(figsize=(10, 6))
sns.histplot(Apple_df['Daily Return'].dropna(), bins=50, kde=True, color='skyblue')
plt.title('Apple Daily Returns Distribution')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Apple Closing Price with 50-Day Rolling Mean

```python
Apple_df['50-Day MA'] = Apple_df['Close ($)'].rolling(window=50).mean()

plt.figure(figsize=(12, 6))
plt.plot(Apple_df['Date'], Apple_df['Close ($)'], label='Close Price', color='blue', alpha=0.7)
plt.plot(Apple_df['Date'], Apple_df['50-Day MA'], label='50-Day MA', color='red')
plt.title('Apple Closing Price with 50-Day Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
